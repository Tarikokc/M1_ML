## Suite de Fibonaci

```python
def fibo(n) : 
    if n <= 1 : 
        return n
    else : 
        return fibo(n-1) + fibo(n-2)
print(fibo(0))
```

```python
def fibo(n) : 
    fibonnaci = { 0 : 0, 1 : 1 }
    
    def fibo_dict(c) :
        if c in fibonnaci :
            return fibonnaci[c]

        value = fibo(c-1) + fibo(c-2)
        fibonnaci[c] = value 
        return value

    result = fibo_dict(n)
    print(fibonnaci) 
    return result 
    
print(fibo(13))
        
```

```python
def trouver_index_zero(L):
    a = 0
    b = len(L) - 1
    
    while a <= b:
        h = (a + b) // 2
        if L[h] == 0:
            return h
        elif L[h] > 0:
            b = h - 1
        else: 
            a = h + 1
    
    return -1  

liste_exemple = [-5, -3, -1, 0, 2, 4, 6]
resultat = trouver_index_zero(liste_exemple)
print(f"\nRésultat final : L'index de 0 est {resultat}")

liste_exemple = [-5, -3, -2,-1,0, 2, 4, 6]
resultat = trouver_index_zero(liste_exemple)
print(f"\nRésultat final : L'index de 0 est {resultat}")
```
