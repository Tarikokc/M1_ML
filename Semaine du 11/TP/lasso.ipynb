# TP Lasso & Ridge - RÃ©gression  

```python
#importer vos libs 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler

import warnings
warnings.simplefilter("ignore")

df = pd.read_csv("../data/houseData.csv")
print("Dimensions du dataset:", df.shape)

```

```python
columns_to_drop = ['id', 'date', 'waterfront', 'view']
df = df.drop(columns_to_drop, axis = 1)
print("Nouvelles dimensions:", df.shape)
```

```python
sns.pairplot(df[['price', 'bedrooms', 'bathrooms', 'sqft_living']])
plt.show()
```

```python
y = df['price'].values
print("Taille du vecteur cible :", len(y))
```

```python
X = df.drop('price', axis = 1)
print("\nColonnes utilisÃ©es pour la prÃ©diction:")
print(X.columns.tolist())
```

```python
lasso = Lasso(
    alpha=0.2,
)
lasso.fit(X, y)
```

```python
print("Coefficients avec alpha=0.2:")
print(lasso.coef_)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(len(lasso.coef_)), lasso.coef_, 'bo-', linewidth=2, markersize=8)
plt.xticks(range(len(X.columns)), X.columns, rotation=45)
plt.grid(True)
plt.title('Coefficients Lasso avec alpha=0.2')
plt.ylabel('Valeur des coefficients')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)  # Ligne Ã  y=0
plt.tight_layout()
plt.show()
```

```python
lasso_2 = Lasso(alpha=10)
lasso_2.fit(X, y)
print("\nCoefficients avec alpha=1:")
print(lasso_2.coef_)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(len(lasso_2.coef_)), lasso_2.coef_, 'go-', linewidth=2, markersize=8)
plt.xticks(range(len(X.columns)), X.columns, rotation=45)
plt.grid(True)
plt.title('Coefficients Lasso avec alpha=1')
plt.ylabel('Valeur des coefficients')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.1)
plt.tight_layout()
plt.show()
```

```python
lasso_3 = Lasso(alpha=1000)
lasso_3.fit(X, y)
print("\nCoefficients avec alpha=1000:")
print(lasso_3.coef_)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(len(lasso_3.coef_)), lasso_3.coef_, 'ro-', linewidth=2, markersize=8)
plt.xticks(range(len(X.columns)), X.columns, rotation=45)
plt.grid(True)
plt.title('Coefficients Lasso avec alpha=1000')
plt.ylabel('Valeur des coefficients')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
