```python
#importer vos libs 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge

import warnings
warnings.simplefilter("ignore")

df = pd.read_csv("../data/houseData.csv")
print("Dimensions du dataset:", df.shape)

```

```python
columns_to_drop = ['id', 'date', 'waterfront', 'view']
df = df.drop(columns_to_drop, axis = 1)
print("Nouvelles dimensions:", df.shape)
```

```python
sns.pairplot(df[['price', 'bedrooms', 'bathrooms', 'sqft_living']])
plt.show()
```

```python
y = df['price'].values
print("Taille du vecteur cible :", len(y))
```

```python
X = df.drop('price', axis = 1)
print("\nColonnes utilisÃ©es pour la prÃ©diction:")
print(X.columns.tolist())
```

```python
lasso = Lasso(
    alpha=0.2,
    copy_X=True,
    fit_intercept=True,
    max_iter=1000,
    positive=False,
    precompute=False,
    random_state=None,
    selection='cyclic',
    tol=0.0001,
    warm_start=False
)
lasso.fit(X, y)
```

```python
print("Coefficients avec alpha=0.2:")
print(lasso.coef_)
```
